```python
import pandas as pd
import numpy as np
import seaborn as sns
from lin_reg.LinearRegression import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
training = pd.read_csv("DS/swedish_insurance.csv")
```

```python
X_train = training["X"]
y_train = training["Y"]
```

```python
X_train = np.asanyarray(X_train).reshape(-1,1)
```

```python
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
```

```python
lr = LinearRegression(X_train, y_train)
```

```python
lr.train(learning_rate=0.01, epochs=1000)
```

```python
y_pred = lr.predict(X_train)
```

```python
lr.accuracy(X_train, y_train, error="r2")
```

```python
best_w, best_b = lr.best_acc(X_train, y_train)
```

```python
y_pred = lr.predict(X_train, best_w, best_b)
```

```python
import matplotlib.pyplot as plt

plt.scatter(training["X"], y_train, color="blue")
plt.plot(training["X"], y_pred, "r")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
